```python
import numpy as np
import pandas as pd
```

```python
#seleção (filtragem de linhas específicas de um dataframe)
dataframe_01=pd.DataFrame({
    "região":["america do sul", "america do sul", "america do sul", "america do sul","africa","asia"],
    "estados":["brasil","argentina","uruguai","paraguai","angola","timor leste"],
    "idh":[0.754,0.842,0.809,0.724,0.574, 0.607]
                           })
dataframe_01
```

```python
novo = pd.DataFrame({
    "região":"africa",
    "estados":"moçambique",
    "idh": 0.446
},index=[0])
novo
```

```python
inserção = pd.concat([dataframe_01, novo],ignore_index=True,sort=False)
inserção
```

```python
exclusão = inserção.drop([1,5])
exclusão
```

```python
exclusão = exclusão.reset_index(drop=True)
exclusão
```

```python
inserção.at[0,"região"]="AFRICA"
```

```python
inserção.loc[inserção["estados"]=="timor leste","região"]="ASIA"
inserção
```
